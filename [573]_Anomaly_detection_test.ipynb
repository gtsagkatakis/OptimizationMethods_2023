```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt


mat = scipy.io.loadmat('satimage-2.mat') #https://odds.cs.stonybrook.edu/satimage-2-dataset/
#mat = scipy.io.loadmat('cardio.mat')  #https://odds.cs.stonybrook.edu/cardiotocogrpahy-dataset/

```

```python
print(mat.keys())
print(mat['X'])
```

```python
dataX = mat['X']
datay = mat['y']
```

```python
plt.imshow(dataX, cmap='gray')
plt.show()

plt.plot(datay)
plt.show()
```

```python

# Perform Singular Value Decomposition
U, S, V = np.linalg.svd(dataX)

# Plot the singular values
plt.plot(S, marker='o')
plt.title('Singular Values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```
