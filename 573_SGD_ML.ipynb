```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

np.random.seed(1)
```

```python
# Load California Housing dataset
california = fetch_california_housing()
X, y = california.data, california.target

# Preprocess the data: Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

```

```python
# Mini-batch Gradient Descent
def mini_batch_SGD(X, y, learning_rate, num_iterations, batch_size):
    n_samples, n_features = X.shape
    weights = np.zeros(n_features)  # Initialize weights
    bias = 0
    mse_history = []  # To store the MSE values

    for _ in range(num_iterations):
        idx = np.random.randint(0, n_samples, batch_size)
        X_batch, y_batch = X[idx], y[idx]

        # Compute predictions
        y_pred = np.dot(X_batch, weights) + bias
        error = y_batch - y_pred

        # Compute gradients
        grad_weights = -2 * np.dot(X_batch.T, error) / batch_size
        grad_bias = -2 * np.sum(error) / batch_size

        # Update weights and bias
        weights -= learning_rate * grad_weights
        bias -= learning_rate * grad_bias

        # Calculate MSE for this iteration and append to history
        y_train_pred = np.dot(X_train, weights) + bias
        mse = np.mean((y_train - y_train_pred)**2)
        mse_history.append(mse)

    return weights, bias,mse_history


```

```python
# Parameters
learning_rate = 0.01
num_iterations = 1000
batch_size = 1

# Train the model using Mini-batch SGD
weights, bias, mse_history  = mini_batch_SGD(X_train, y_train, learning_rate, num_iterations, batch_size)

# Evaluate the model
def predict(X, weights, bias):
    return np.dot(X, weights) + bias

y_pred = predict(X_test, weights, bias)
mse = np.mean((y_test - y_pred)**2)
#print("Mean Squared Error on Test Set:", mse)

# Plot MSE history
plt.plot(range(num_iterations), mse_history)
plt.title("MSE during Training")
plt.xlabel("Iteration")
plt.ylabel("Mean Squared Error")
plt.title(f"Training Progress - Final MSE: {mse:.4f}")
plt.show()
```
