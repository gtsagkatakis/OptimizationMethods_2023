```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def gram_schmidt(vectors):
    basis = []
    for v in vectors.T:
        w = v - np.sum(np.dot(v, b.T)*b for b in basis)
        if (np.linalg.norm(w) > 1e-10):
            basis.append(w/np.linalg.norm(w))
    return np.array(basis)
```

```python
# Load the iris data
iris = datasets.load_iris()
data = iris.data.T  # Transpose data so each feature vector is a column

# We're considering only the first 3 features for a 3D space representation
data = data[:3]
```

```python
# Perform Gram-Schmidt
orthonormal_basis = gram_schmidt(data)

# Prepare for 3D visualization
fig = plt.figure(figsize=(12, 6))

# 3D scatter plot of the original data
ax = fig.add_subplot(121, projection='3d')
ax.scatter(data[0], data[1], data[2], c=iris.target, cmap='viridis', edgecolor='k')
ax.set_title("Original Data")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_zlabel("Petal Length")

# 3D plot for basis vectors
ax = fig.add_subplot(122, projection='3d')

# Orthonormal basis vectors from Gram-Schmidt
for vector in orthonormal_basis:
    ax.quiver(0, 0, 0, vector[0], vector[1], vector[2], color='b', length=1.0)

ax.set_title("Orthonormal Basis Vectors")
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```
